```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 25)
x
```

```python
y = x * x + 2
y
```

```python
#creating the canvas
fig = plt.figure()

#creating an axes on the canvas
axes = fig.add_axes([0.5, 0.1, 0.8, 0.8])
axes.plot(x, y, 'r*')

#adding label
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('My Plot')
```

```python
#create the canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
#subplots using tuple unpacking
fig, axes = plt.subplots(nrows=1,ncols=2)

#use plt.tight_layout() to fix overlapping
#axes.plot(x,y)

axes[0].plot(x,y)
axes[0].set_xlabel('X label')
axes[0].set_ylabel('Y label')
axes[0].set_title('Plot 1')

axes[1].plot(y,x)
axes[1].set_xlabel('Y label')
axes[1].set_xlabel('Y label')
axes[1].set_title('Plot 2')
```

```python
#figure size
fig = plt.figure(figsize=(10,10))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# fig size in subplots
#create the canvas, specify the size
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(4,4))

axes[0].plot(x,y) 
axes[1].plot(y,x)
```

```python
#saving the fig
fig.savefig('sample.jpg')
```

```python
#adding legend on a plot
fig = plt.figure(figsize=(5,5))

#add an axes to the canvas
axes = fig.add_axes([0, 0, 0.8, 0.8])

axes.plot(x, x**2, label='X squared')
axes.plot(x, x**3, label='X cube')

# in order for the legend to work you need to add the label in the plot
axes.legend()
```

```python
# plot color, linewidth, alpha, etc
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x,y, color='green', linewidth=10)
```

```python
# we can also use RGB Hex code for a more specific color
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(y,x, color='#CC0066', lw=10)
```

```python
#adding linestyle and marker, markersize
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(y,x, color='green', lw=1, ls='--', marker='+', markersize=10)
```
