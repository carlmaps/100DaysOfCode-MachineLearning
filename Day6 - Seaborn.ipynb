```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.datasets import load_boston
```

#### Seaborn Distribution Plots
1. dist plot
2. joint plot
3. pair plot

```python
housing = load_boston()
```

```python
housing.keys()
```

```python
print(housing.DESCR)
```

```python
housing.feature_names
```

```python
boston_data = pd.DataFrame(data=housing.data, columns=housing.feature_names)
```

```python
boston_data.head()
```

```python
boston_data['Price'] = housing.target
boston_data.head()
```

#### Dist Plot

```python
sns.set_style("whitegrid")
with sns.color_palette("BuGn_r"):
    sns.distplot(boston_data['Price'], bins=30)
```

#### Joint Plot

```python
with sns.color_palette("BuGn_r"):
    sns.jointplot(x='Price',y='LSTAT', data=boston_data)
```

```python
sns.set_style("whitegrid")
with sns.color_palette("BuGn_r"):
     sns.jointplot(x='Price',y='LSTAT', data=boston_data, kind='hex')
```

```python
sns.set_style("whitegrid")
with sns.color_palette("BuGn_r"):
     sns.jointplot(x='Price',y='LSTAT', data=boston_data, kind='reg')
```

```python
sns.set_style("whitegrid")
with sns.color_palette("BuGn_r"):
     sns.jointplot(x='Price',y='LSTAT', data=boston_data, kind='kde')
```

#### Pairplot

```python
sns.pairplot(boston_data, palette='coolwarm')
```

#### Seaborn Categorical Plots

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
iris.feature_names
```

```python
iris.target
```

```python
iris_data = pd.DataFrame(data = iris.data, columns=iris.feature_names)
```

```python
iris_data.head()
```

```python
iris_data['Species'] = iris.target
```

```python
iris_data['Species_Names'] = iris_data['Species'].apply(lambda x: iris.target_names[x])
```

```python
iris_data.head()
```

##### Bar plot

```python
#sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='Species', y='sepal length (cm)', data=iris_data, hue='Species_Names')
```

#### Count Plot

```python
sns.countplot('Species', data=iris_data)
```

#### Boxplot

```python
sns.boxplot(x='Species', y='sepal width (cm)', data = iris_data, hue='Species_Names')
```

```python
sns.boxplot(x='Species', y='sepal length (cm)', data = iris_data, hue='Species_Names')
```
